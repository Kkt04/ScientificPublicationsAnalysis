```python

# Cell 1: Import libraries and setup
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sys
import os

# Add src to path so we can import our modules
sys.path.append('../src')

# Setup plotting
plt.style.use('default')
sns.set_palette("husl")
print("‚úÖ Libraries imported successfully!")

# Cell 2: Import our custom modules
from data_loader import DataLoader
from analyzer import PublicationAnalyzer

print("‚úÖ Custom modules imported successfully!")

# Cell 3: Load and explore data
loader = DataLoader()
data = loader.create_sample_from_real_data(sample_size=1000)

print("üìä Dataset Overview:")
print(f"Shape: {data.shape}")
print(f"Columns: {list(data.columns)}")

# Cell 4: Display first few rows
print("\nüëÄ First 5 rows:")
data.head()

# Cell 5: Basic statistics
print("\nüìà Basic Statistics:")
print(data.describe())

# Cell 6: Category analysis
print("\nüî¨ Category Analysis:")
categories = data['categories'].str.split().explode()
print(f"Total unique categories: {categories.nunique()}")
print("\nTop 10 categories:")
top_cats = categories.value_counts().head(10)
print(top_cats)

# Cell 7: Simple visualization
plt.figure(figsize=(10, 6))
top_cats.plot(kind='bar', color='lightblue')
plt.title('Top 10 Research Categories in Our Sample')
plt.xlabel('Category')
plt.ylabel('Number of Papers')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Cell 8: Timeline analysis
data['year'] = pd.to_datetime(data['update_date']).dt.year
yearly_counts = data['year'].value_counts().sort_index()

plt.figure(figsize=(10, 6))
yearly_counts.plot(kind='line', marker='o')
plt.title('Publications Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Papers')
plt.grid(True, alpha=0.3)
plt.show()
print(" Exploration complete! Ready for deeper analysis.")
```
